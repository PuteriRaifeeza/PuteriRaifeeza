```python
import pandas as pd
df = pd.read_csv("drug200.csv")
df.head()
```

```python
X = df.iloc[:,:-1]
X.head()
#panggil X value
```

```python
y = df['Drug']
df.head()
```

```python
#convert DF to array
type(X)
X=X.values
X[:5]
```

```python
#convert categorical to numeric
from sklearn.preprocessing import LabelEncoder
label_sex = LabelEncoder()
X[:,1]=label_sex.fit_transform(X[:,1])
label_BP= LabelEncoder()
X[:,2]=label_sex.fit_transform(X[:,2])
label_chole = LabelEncoder()
X[:,3]=label_sex.fit_transform(X[:,3])
```

```python
#convert categorical to numeric
from sklearn.preprocessing import LabelEncoder

label_sex = LabelEncoder()
# Use .iloc to select rows and columns by their integer positions
X.iloc[:,1]=label_sex.fit_transform(X.iloc[:,1])
label_BP= LabelEncoder()
X.iloc[:,2]=label_sex.fit_transform(X.iloc[:,2])
label_chole = LabelEncoder()
X.iloc[:,3]=label_sex.fit_transform(X.iloc[:,3])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
X_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import jaccard_score

DT = DecisionTreeClassifier() #criterion="entropy", max_depth = 4

#training
DT.fit(X_train, y_train) # it shows the default parameters

#prediction
y_hat = DT.predict(X_test)

#evaluation
print("The DecisionTrees's Accuracy is: ", accuracy_score(y_test, y_hat))
print("The confusion matrix is: ", confusion_matrix(y_test, y_hat))
print("The jaccard score is: ", jaccard_score(y_test, y_hat, average ='micro'))
```
